###  5. Используя pandas, проведи когортный анализ пользователей. В период с января по декабрь выяви когорту с самым высоким retention на 3й месяц. Описание подхода можно найти тут. (10 баллов)

```python
#создаем датафрейм
orders_uniq_id = olist_orders.merge(olist_customers, how='inner', on='customer_id')
orders_uniq_id.head()
```

```python
#оставляем заказы со статусом 'delivered'
delivered_orders_uniq_id = orders_uniq_id[orders_uniq_id.order_status == 'delivered']
delivered_orders_uniq_id.head()
```

```python
#переводим формат колонки даты в месяц
delivered_orders_uniq_id.order_purchase_timestamp = delivered_orders_uniq_id.order_purchase_timestamp.dt.to_period('M')
```

```python
#оставляем только нужные колонки и сортируем по дате
delivered_orders_uniq_id = delivered_orders_uniq_id[['customer_unique_id','order_id','order_purchase_timestamp']]
delivered_orders_uniq_id = delivered_orders_uniq_id.sort_values('order_purchase_timestamp')
delivered_orders_uniq_id.head()
```

```python
#создаем колонку с датой первого заказа 
first_date = delivered_orders_uniq_id.groupby('customer_unique_id',as_index=False) \
    .agg({'order_purchase_timestamp':'first'})
first_date.head()
```

```python
#объединяем основной датасет с колонкой даты первого заказа
delivered_orders_uniq_id = delivered_orders_uniq_id.merge(first_date, how='inner',on='customer_unique_id')
```

```python
#переименуем название колонок для удобства
delivered_orders_uniq_id = delivered_orders_uniq_id \
.rename(columns={'order_purchase_timestamp_x':'order_date','order_purchase_timestamp_y':'first_order_date'})
delivered_orders_uniq_id.head()
```

```python
delivered_orders_uniq_id = delivered_orders_uniq_id.groupby(['first_order_date','order_date']) \
    .agg({'customer_unique_id':'count'})
```

```python
#добавляем функцию,считающую когорту
def cohort_period(df):
    df['cohort_period'] = np.arange(len(df)) + 1
    return df

delivered_orders_uniq_id = delivered_orders_uniq_id.groupby(level=0).apply(cohort_period)
delivered_orders_uniq_id.head()
```

### 6. Часто для качественного анализа аудитории используют подходы, основанные на сегментации. Используя python, построй RFM-сегментацию пользователей, чтобы качественно оценить свою аудиторию. В кластеризации можешь выбрать следующие метрики: R - время от последней покупки пользователя до текущей даты, F - суммарное количество покупок у пользователя за всё время, M - сумма покупок за всё время. Подробно опиши, как ты создавал кластеры. Для каждого RFM-сегмента построй границы метрик recency, frequency и monetary для интерпретации этих кластеров. Пример такого описания: RFM-сегмент 132 (recency=1, frequency=3, monetary=2) имеет границы метрик recency от 130 до 500 дней, frequency от 2 до 5 заказов в неделю, monetary от 1780 до 3560 рублей в неделю.

```python
#объединяем датафреймы
olist_orders_uniq_rfm = olist_orders.merge(olist_customers, on='customer_id', how='left')
orders_uniq_id_rfm = olist_orders_uniq_rfm.merge(olist_order_items, how='left', on='order_id')
orders_uniq_id_rfm.head()
```

```python
#оставляем только выполненые заказы
orders_uniq_id_rfm = orders_uniq_id_rfm[orders_uniq_id_rfm.order_status == 'delivered']
```

```python
#оставляем необходимые колонки
orders_uniq_id_RFM = orders_uniq_id_rfm[['customer_unique_id','order_id','order_purchase_timestamp','price']]
orders_uniq_id_RFM.head()
```

```python
#переводим данные в нужный формат
orders_uniq_id_RFM.order_purchase_timestamp = pd.to_datetime(orders_uniq_id_RFM.order_purchase_timestamp)
```

```python
# считаем разницу между датой нынешней и датой последнего заказа
orders_uniq_id_RFM['time_delta'] = orders_uniq_id_RFM.order_purchase_timestamp \
    - orders_uniq_id_RFM.order_purchase_timestamp.max()
orders_uniq_id_RFM.head()
```

```python
orders_uniq_id_RFM['r_score'] = pd.qcut(orders_uniq_id_RFM['time_delta'],q = [0,0.8,0.9,1] , labels=[1,2,3])
#сегментируем r-score поделив на 3 значения до 30 дней, до 96 дней, более 96 дней
orders_uniq_id_RFM.head()
```

```python
#объединяем результат r-score с итоговой колонкой
olist_orders_uniq_id = olist_orders.merge(olist_customers, on='customer_id', how='left') 
olist_orders_uniq_id.head()
```

```python
#считаем кол-во выполненых заказов для каждого пользователя
x = olist_orders_uniq_id.loc[olist_orders_uniq_id['order_status'] == 'delivered'] \
    .groupby('customer_unique_id',as_index=False) \
    .agg({'order_id':'count'})
x.head()
```

```python
#убираем повторенные значения
orders_uniq_id_RFM.customer_unique_id = orders_uniq_id_RFM.customer_unique_id.drop_duplicates()
```

```python
#убираем пропущенные данные
orders_uniq_id_RFM = orders_uniq_id_RFM.dropna()
```

```python
#объединяем результат с итоговой колонки
orders_uniq_id_RFM = orders_uniq_id_RFM.merge(x,how='left', on='customer_unique_id')
orders_uniq_id_RFM.head()
```

```python
#сегментируем f-score, 1 если 1 заказ, 2 если 2 заказа , 3 если более 2 заказов
orders_uniq_id_RFM['f_score'] = orders_uniq_id_RFM.order_id_y.apply(lambda x: 3 if x > 2 else x)
orders_uniq_id_RFM.head()
```

```python
#проверяем результаты 
orders_uniq_id_RFM.f_score.value_counts()
```

```python
#отбираем выполненые заказы
delivered_order_id = olist_orders[olist_orders['order_status'] == 'delivered']
```

```python
#считаем суммарный заработок на пользователях
y = orders_uniq_id_rfm.groupby('customer_unique_id',as_index=False) \
    .agg({'price':'sum'})
y.head()
```

```python
#сегментируем 3 группа sum  154, 2 группа от 89.73 до 154 , остальное 1 группа
y['m_score'] = y.price.apply(lambda x: 3 if x > 154 else 2 if 154 > x and x > 89.73 else 1)
y.head()
```

```python
#объединяем m-score с основным датасетом
orders_uniq_id_RFM = orders_uniq_id_RFM.merge(y, how='left', on='customer_unique_id')
orders_uniq_id_RFM.head()
```

```python
#оставляем нужные колонки
orders_uniq_id_RFM = orders_uniq_id_RFM[['customer_unique_id','r_score','f_score','m_score']]
orders_uniq_id_RFM.head()
```

```python
orders_uniq_id_RFM.m_score.value_counts()
```

```python
#итоговый RFM показатель
orders_uniq_id_RFM['RFM'] = orders_uniq_id_RFM.r_score.astype(str) \
    +orders_uniq_id_RFM.f_score.astype(str) \
    +orders_uniq_id_RFM.m_score.astype(str)
orders_uniq_id_RFM.head()
```

```python
#готовим данные для результирующей таблицы
RFM_results = orders_uniq_id_RFM.groupby('RFM',as_index=False) \
    .agg({'customer_unique_id':'count'})
RFM_results.head()
```

```python
plt.figure(figsize=(18,14))
plt.title('RFM')
sns.barplot(data = RFM_results , x='RFM', y= 'customer_unique_id')
```

```python
RFM_results = RFM_results.rename(columns={'customer_unique_id':'count'}) \
    .sort_values('count',ascending=False)
RFM_results['percent'] = round((RFM_results['count'] * 100) / RFM_results['count'].sum(),2).astype(str) +' %'
RFM_results
```

### Немного информации о значения RFM:

### Аббревиатура RFM расшифровывается:

#### - Recency — давность (как давно ваши пользователи что-то у вас покупали);
#### - Frequency — частота (как часто они у вас покупают);
#### - Monetary — деньги (общая сумма покупок).
#### По давности заказа (recency):

#### 1 — давние клиенты;
#### 2 — относительно недавние клиенты;
#### 3 — недавние клиенты.
#### По частоте покупок (frequency):

#### 1 — покупает очень редко (единичные заказы);
#### 2 — покупает нечасто;
#### 3 — покупает часто.
#### По сумме покупок (monetary):

#### 1 — маленькая сумма;
#### 2 — средняя сумма;
#### 3 — большая сумма.

### Например, пользователь «111» покупал давно, один раз и на маленькую сумму. Так себе клиент, иным словом. Или пользователь «333»: покупает часто, на большую сумму и последняя покупка была недавно. Это наши лучшие клиенты.

### Основная доля клиентов состоит в группе RFM 111 ( последняя покупка более 100 дней назад, кол-во покупок = 1, суммарный объем покупки менее 89.73) Мы имеем дело с минимальными показателями retention, и невысоким средним чеком.) Это почти 40 % наших клиентов. Клиенты RFM 333 составляет около 0.01 %
